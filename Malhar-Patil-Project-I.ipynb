**Step 1: Write a function that can print out a board. Set up your board as a list, where each index 1-9 corresponds with a number on a number pad, so you get a 3 by 3 board representation.**

```python
from IPython.display import clear_output #  clear_output is used to ``clear`` the in jupyter notebook. If you are using other IDE, you need to use a different function(google it!)

def display_board(board):
    print(board[1] + ' | ' + board[2] + ' | ' + board[3])
    print('---------')
    print(board[4] + ' | ' + board[5] + ' | ' + board[6])
    print('---------')
    print(board[7] + ' | ' + board[8] + ' | ' + board[9])
```

**Step 2: Write a function that can take in a player input and assign their marker as 'X' or 'O'. Think about using *while* loops to continually ask until you get a correct answer.**

```python
def player_input():
    inp = ''
    while inp != 'X' and inp != 'O':
        inp = input("Choose either 'X' or 'O': ")
    return inp
```

**Step 4: Write a function that takes in a board and a mark (X or O) and then checks to see if that mark has won.**

```python
def win_check(board, mark):
    if board[1] == board[2] == board[3] == mark or board[4] == board[5] == board[6] == mark or board[7] == board[8] == board[9] == mark:
        return True
    elif board[1] == board[4] == board[7] == mark or board[2] == board[5] == board[8] == mark or board[3] == board[6] == board[9] == mark:
        return True
    elif board[1] == board[5] == board[9] == mark or board[3] == board[5] == board[7] == mark:
        return True
    else:
        return False
```

**Step 5: Write a function that uses the random module to randomly decide which player goes first. You may want to write a fucntion similar to flipping a fair coin that you did in Exam1 or  lookup random.randint(). Return a string of which player went first.**

```python
import random

def choose_first(player1, player2):
    p1 = 0
    p2 = 1
    rand = random.randint(0, 1)
    if(rand == p1):
        return player1
    else:
        return player2
```

**Step 6: Write a function that returns a boolean indicating whether a space on the board is freely available.**

```python
def space_check(board, position):
    if board[position] == ' ':
        return True
    else:
        return False
```

**Step 7: Write a function that checks if the board is full and returns a boolean value. True if full, False otherwise.**

```python
def full_board_check(board):
    full = True
    i = 1
    while i < len(board):
        if space_check(board, i):
            full = False
            break
        i += 1
    return full
```

**Step 8: Write a function that asks for a player's next position (as a number 1-9) and then uses the function from step 6 to check if it's a free position. If it is, then return the position for later use.**

```python
def player_choice(board):
    inp = 0
    while space_check(board, inp) == False:
        inp = int(input("Next position (1-9)? "))
    return inp
```

**Step 9: Write a function that asks the player if they want to play again and returns a boolean True if they do want to play again.**

```python
def replay():
    playagain = ' '
    while playagain.lower() != 'y' and playagain.lower() != 'n':
        playagain = input("Do you want to play again (Y/N)? ")
    if playagain.lower() == 'y':
        return True
    else:
        return False
```

**Step 10: Here comes the hard part! Use while loops and the functions you've made to run the game!**

```python
print('Welcome to Tic Tac Toe!')

p1 = input("Player: What is your name? ")
p2 = input("Player: What is your name? ")
i = 1 # Counter to check whether it is the odd or even round

while True:
    # Set the game up here
    board = ['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']
    display_board(board)
    if i % 2 == 1: # Alternate player1 each round
        player1 = choose_first(p1, p2) 
        print(player1 + " goes first")
        player2 = ' '
        if player1 == p1:
            player2 = p2
        else:
            player2 = p1
    elif i % 2 == 0: # Alternate player1 each round
        temp = player1
        player1 = player2
        player2 = temp
        print(player1 + " goes first")
    player1_input = player_input()
    player2_input = ' '
    if player1_input == 'X':
        player2_input = 'O'
    else:
        player2_input = 'X'
        
    
    while win_check(board, 'X') == False or win_check(board, 'O') == False or full_board_check(board) == False:
        #Player 1 Turn
        print("It is " + player1 + "'s turn")
        pos = player_choice(board)
        place_marker(board, player1_input, pos)
        display_board(board)

        if win_check(board, player1_input) == True:
            winner = player1 + " is the winner"
            break
        if full_board_check(board) == True:
            winner = "Board is full"
            break
        
        # Player2's turn.
        print("It is " + player2 + "'s turn")
        pos = player_choice(board)
        place_marker(board, player2_input, pos)
        display_board(board)
        if win_check(board, player2_input) == True:
            winner = player2 + " is the winner"
            break
        if full_board_check(board) == True:
            winner = "Board is full"
            break
    print("Game over. " + winner)
    i += 1 #Counter to check whether this is the odd or even round

    if replay() == False:
        break
```
